```python
import numpy as np
```

```python
def std(x): return np.std(x)
```

```python
HeroInfo.groupby(['Name','FirstAppearance']).agg([std]).head()
```

```python
def nanstd(x): return np.nanstd(x)
```

```python
df = pd.DataFrame({'grp': [0, 0, 1, 1], 'returns': [0, 0.1, -.1, .1]})
```

```python
df.prod()
```

```python
def compound(self, axis=None, skipna=None, level=None):
            if skipna is None:
                skipna = True
            return (1 + self).prod(axis=axis, skipna=skipna, level=level) - 1
```

```python
df.compound()
```

```python
df_1 = 1+df
```

```python
df_1
```

```python
df_1.prod()
```

```python
df_1.prod() -1 
```

```python
df.compound()
```

```python
df['returns'].compound()
```
