```python
asia_dep_port = asia.groupby(["항공사","출발 공항","출발 국가","도착 공항","도착 국가","년도","월"])
```

```python
import numpy as np
```

```python
a = asia_dep_port.agg({'월':'count','년도':'count','출발 국가':'count', '도착 국가':'count','정원(전체)':'sum'}) 
```

```python
a.head(20)
```

```python
a.columns
```

```python
a.columns = ['운항회수', '년도', '출발 국가', '도착 국가', '정원(전체)']
```

```python
a.head()
```

```python
a.drop(columns="년도",axis=1,inplace=True)
```

```python
a.head()
```

```python
a.to_csv("example_lee.csv",encoding='cp949')
```
