## 예제 9-4  plot 그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2016', periods=1000))
```

```python
ts = ts.cumsum()
```

```python
ts.head()
```

```python
ts.plot()
plt.show()
```

```python
ts.plot
```

```python
for i in dir(ts.plot) :
    if not i.startswith("_") :
        print(i, end=",  ")
```

```python
ts.plot.line
```

```python
ts.plot.line.__func__
```

```python
ts.plot.line()
```

```python
ts.plot(kind='line')
```

```python
ts.index[:10]
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), 
                  index=ts.index, 
                  columns=list('가나다라'))
```

```python
df = df.cumsum()
```

```python
df.head()
```

```python
df.plot(title="한글")
plt.show()
```

```python
df.plot.line().legend(['A_가','B_나','C_다','D_라'])
```

```python
df.plot(x=df.index, y='나')
```

```python
df.plot.line(x=df.index, y='나')
```
